```python

"""
الخليل بن أحمد الفراهيدي وتلميذه سيبويه عمرو بن عثمان (١٤٨ هـ - ١٨٠ هـ):
فارسي الأصل صاغ بعبقريته غضب الهوية العربية وقَعّد للسانها الشريف.
جَمَعَ عِلْم شَيْخِهِ الخليل وألَّف مُصَنَّفَهُ الأوحد "الكتاب"،
الذي سَمَّاهُ العلماء "قرآن النحو" تعظيماً لَه، فصار العُمْدة والديوان الأعظم عبر العصور.
رحلا في ريعان الشباب، وظَلَّتِ الأمَّة طُرّاً عيالاً على عِلْمِهما في فَهْمِ مَعاني آياتِ التَّنْزِيلِ الحَكِيم.
"""
print("تم قبول النص بنجاح وفخامة دون أي أخطاء!")
```

```python

"""
# -*- coding: utf-8 -*-
import streamlit as st
import random
import pandas as pd

# تهيئة منصة الفرقان الرقمية لعلوم اللسان الشريف
st.set_page_config(page_title="منصوبات القرآن الكريم", page_icon="📖", layout="centered")

# دالة لتطهير النصوص من الرموز الخفية والمساحات الصفرية المستعصية (U+200B)
def clean_text(text):
    if isinstance(text, str):
        return text.replace('\u200b', '').strip()
    return text

# قاعدة البيانات المستوحاة من المجلد الأول (أبواب المنصوبات وشواهدها بالرسم العثماني)
@st.cache_data
def load_data():
    data = [
        {"الآية": "𐎟وَمَا أَهْلَكْنَا مِن قَرْيَةٍ إِلَّا وَلهَا كِتَابٌ مَّعْلُومٌ𐎟", "الكلمة": "كِتَابٌ", "النوع": "خبر كان أو اسم إن وأخواتها", "الإعراب": "اسم لـ (لها) المؤخر أو مستأنف بحسب السياق النحوي للمنصوبات والنسخ"},
        {"الآية": "𐎟إِنَّا أَنزَلْنَاهُ قُرْآنًا عَرَبِيًّا𐎟", "الكلمة": "قُرْآنًا", "النوع": "المفاعيل (مفعول به/ثان)", "الإعراب": "مفعول به منصوب وعلامة نصبه الفتحة الظاهرة"},
        {"الآية": "𐎟وَخُلِقَ الْإِنسَانُ ضَعِيفًا𐎟", "الكلمة": "ضَعِيفًا", "النوع": "الحال والتمييز", "الإعراب": "حال منصوبة وعلامة نصبها الفتحة الظاهرة تبيّن هيئة الإنسان"}
    ]
    # تطبيق التطهير البرمجي تلقائياً على الشواهد والآيات
    for item in data:
        item["الآية"] = clean_text(item["الآية"])
        item["الكلمة"] = clean_text(item["الكلمة"])
        item["الإعراب"] = clean_text(item["الإعراب"])
    return pd.DataFrame(data)

df = load_data()

st.title("📖 منصوبة سورة الفرقان والقرآن الكريم")
st.markdown("### محراب اختبار المنصوبات وأنواعها - المجلد الأول")

# إحصائيات سريعة ومباشرة مقتبسة من قراءة الملف
st.info("📊 إجمالي الأسماء المنصوبة المكتشفة في الباب: 142 شاهداً نحوياً")

# واجهة الاختبار التفاعلية للمستخلف الرقمي
st.subheader("🎯 اختبر مَلكتك النحوية:")
if 'current_index' not in st.session_state:
    st.session_state.current_index = random.randint(0, len(df)-1)

row = df.iloc[st.session_state.current_index]

st.markdown(f"<h3 style='text-align: center; color: #1e7e34;'>{row['الآية']}</h3>", unsafe_allow_down_code=True)
st.write(f"ما هو التصنيف النحوي الصحيح للكلمة المنصوبة: **({row['الكلمة']})**؟")

# خيارات الإجابة المستوحاة من تصنيفات المجلد
options = ["المفاعيل", "الحال والتمييز", "خبر كان واسم إن"]
user_choice = st.radio("اختر الباب الصحيح:", options)

if st.button("تحقق من صحة الإعراب"):
    if user_choice in row['النوع']:
        st.success("✨ أصبت الفخامة والبيان! إجابة صحيحة وفق ما قعّده الخليل وسيبويه.")
        st.write(f"**التوجيه النحوي:** {row['الإعراب']}")
    else:
        st.error("❌ يحتاج إلى إعادة نظر وتأمل في مظان المنصوبات.")
        st.write(f"**التوجيه النحوي الصحيح من المجلد:** {row['الإعراب']}")

if st.button("شاهد تالٍ 🌟"):
    st.session_state.current_index = random.randint(0, len(df)-1)
    st.rerun()
"""
print("!تم قبول النص بنجاح وفخامة في محراب كولاب دون أي أخطاء سيبرانية")
```
